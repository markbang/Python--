```python
import seaborn as sns
```

# 1.关系绘图

```python
# replot基本使用
%matplotlib inline
tips = sns.load_dataset('tips', cache=True)
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
# hue参数  hue参数可以用来区分不同的类别，显示不同颜色
sns.relplot(x='total_bill', y='tip', hue='smoker', data=tips)
```

```python
# 添加col和row参数，可以将hue参数的类别分别绘制在不同的图中
sns.relplot(x='total_bill', y='tip', hue='smoker', col='time', data=tips)
```

```python
# 也可以再在row中添加一个类别
sns.relplot(x="total_bill",y="tip",hue="day",col="time",row="sex",data=tips)
```

```python
# 有时候我们的图有很多，默认情况下会在一行中全部展示出来，那么我们可以通过col_wrap来指定具体多少列
sns.relplot(x="total_bill",y="tip",col="day",col_wrap=3,data=tips)
```

```python
# 绘制折线图
fmri = sns.load_dataset('fmri', cache=True)
sns.relplot(x='timepoint', y='signal', kind='line', data=fmri)
```

```python
# 设置hue为event，就会根据event来绘制不同的颜色
# 设置col为region，就会根据region值的个数来绘制指定个数的图
# 设置style为event，就会根据event来设置线条的样式
sns.relplot(x="timepoint",y="signal",kind="line",hue="event",col="region",style="event",data=fmri)
```

## 2.1 分类散点图

```python
# catplot基本使用
sns.catplot(x="day",y="total_bill",data=tips,hue="sex")# 默认情况下，kind='strip'
```

```python
sns.catplot(x="day",y="total_bill",kind="swarm",data=tips,hue="sex")# kind='swarm'，可以避免散点重叠
```
